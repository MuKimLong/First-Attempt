```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# Barplots

## Set darkgrid style from seaborn

```python
sns.set_style("darkgrid")
```

## Make a bar plot

```python
df.plot.bar()
```

```python
df.plot(kind = "bar", figsize =(10,14))
```

## Annotate the hight of the bars on top of them

```python
ax = df.plot.bar(figsize = (9,6), rot=0)
sns.set_style(style = "darkgrid")
plt.show()
```

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]

g = df.plot.bar(figsize=(9,6), rot=0)
g.set_xticklabels(labels)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.02, p.get_height()+0.5))
```

## Stack the bars

```python
ax = df.plot.bar(stacked=True, ax=ax, rot=0)
plt.figure(figsize=(8,5))
for bar in ax.patches:
    ax.annotate("%" + str(round(100*bar.get_height()/len(df), 1)) + ' - ' + str(bar.get_height() ),
                (bar.get_x()+0.2, bar.get_height()+300),
                size=12)
```

```python
ax = df.plot.bar(stacked = True)
sns.set_style(style = "darkgrid")
plt.xticks(rotation=90)
plt.show()
```

## Rotate the labels and set figsize

```python
df.plot.bar(stacked=True, rot = 0, figsize=(9,6));
```

```python
ax = df.plot.bar(figsize = (9,6), rot=0, stacked = True)
sns.set_style(style = "darkgrid")
plt.show()
```

## Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
# without using df.set_axis() function:
fig, ax=plt.subplots(figsize=(9,6))
df.plot(kind='bar', stacked=True, ax=ax, rot=0)
plt.xticks(np.arange(7), ['Game-1', 'Game-2', 'Game-3', 'Game-4', 'Game-5', 'Game-6', 'Game-7']);
```

```python
df = df.set_axis(["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"])
df
```

```python
ax = df.plot.bar(figsize = (9,6), rot=0, stacked = True)
sns.set_style(style = "darkgrid")
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(9,6))
g = df.plot(kind='bar', stacked=True, ax=ax, rot=0, )
g
plt.xticks(np.arange(7), ['Game-1', 'Game-2', 'Game-3', 'Game-4', 'Game-5', 'Game-6', 'Game-7']);
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.08, p.get_height()+0.5))
```

# Histograms

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

## Histogram of the total_bill column

```python
tips.total_bill.plot.hist(bins=10)
```

```python
sns.distplot(tips.total_bill, bins=10, kde=True)
```

```python
plt.hist(tips.total_bill)
plt.show()
```

# Lineplots

```python
tips.head()
```

## Plot the avg tip by size  (lineplot)

```python
tips.groupby('size').tip.mean().plot()
```

```python
sns.lineplot(data=tips, x="size", y="tip", ci=None)
```

```python
df1= tips.groupby('size')['tip'].mean()
df1
```

```python
fig, ax = plt.subplots()
ax.plot(df1.index, df1.values)
ax.set_xlabel("size");
```

## Set the linestyle as "--"

```python
tips.groupby('size').tip.mean().plot(ls=":")
```

```python
sns.lineplot(data=tips, x="size", y="tip", ci=None, ls="-.")
```

# Scatter Plots

```python
tips.head()
```

## Make a scatter plot between tip and total_bill

```python
tips.plot.scatter("tip", "total_bill")
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 3)
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```

```python
tips.plot.scatter(x = "tip", y = "total_bill")
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, s=60)
```

## Set an additional dimension using size column

```python
tips.plot.scatter("tip", "total_bill", c="size", cmap="coolwarm")
```

```python
tips.plot.scatter(x = "tip", y = "total_bill", c = tips["size"], cmap = 'coolwarm')
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001, c = tips["size"], cmap='coolwarm')
plt.colorbar()
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```

```python
sns.scatterplot(x='tip', y='total_bill', hue='size', data= tips, palette='coolwarm',legend="full")
```

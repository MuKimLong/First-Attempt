```python
import pandas as pd
```

```python
data1 = [{'name': 'JOHN', 'age': 27}, {'name': 'HENRY2', 'age': 14}]

data2 = { 1:{'name': 'vikash', 'age': 27}, 2:{'name': 'Satyam', 'age': 14} }

data3 = [{'id': 1, 'name': "Cole Volk", 'fitness': {'height': 130, 'weight': 60}}, {'name': "Mose Reg", 'fitness': {'height': 130, 'weight': 60}}, {'id': 2, 'name': 'Faye Raker', 'fitness': {'height': 130, 'weight': 60}}]
```

```python
print(data1)
print(data2)
print(data3)
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2).T
df3 = pd.DataFrame.from_dict(data3)
```

```python
print(df1)
print(df2)
print(df3)
```

```python
df1_2 = df1.append(df2, ignore_index =True)
df1_2
```

```python
df3["height"] = df3.fitness[0]["height"]
df3["weight"] = df3.fitness[0]["weight"]
```

```python
df3
```

```python
df3  = df3.drop(columns=["id", "fitness"])
df3
```

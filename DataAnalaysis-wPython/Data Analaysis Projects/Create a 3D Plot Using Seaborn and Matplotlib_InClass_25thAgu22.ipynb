```python
sns.set_style("whitegrid")  # seaborn

OMEGA = 2

Z1 = np.linspace(0, 10, 100)
X1 = np.cos(OMEGA*Z1)
Y1= np.sin(OMEGA*Z1)

plt.figure(figsize=(6,5))
axes = plt.axes(projection='3d')
axes.plot3D(X1,Y1,Z1)

plt.tight_layout()
plt.show()
```

```python
def FUNC_Z(x, y):
    return 50 - (x**2 + y**2)


sns.set_style('whitegrid')  # seaborn
N = 50

X_VAL = np.linspace(-5, 5, N)
Y_VAL = np.linspace(-5, 5, N)

X1, Y1 = np.meshgrid(X_VAL, Y_VAL)
# mashgrid: Return coordinate matrices from coordinate vectors.

Z1 = FUNC_Z(X1, Y1)

axes = plt.axes(projection='3d')
axes.plot_surface(X1, Y1, Z1)
# axes.plot_wireframe(X1, Y1, Z1)  # plot_surface yerine bunu kullanabilirsiniz.
                                   # tel kafes biçiminde kaplama yapar.
plt.show()
```
